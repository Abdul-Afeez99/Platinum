```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine 'title' and 'news_url' columns to create a new text column
news_df['text'] = news_df['title'] + ' ' + news_df['news_url']

# Replace NaN values with an empty string in the 'text' column
news_df['text'].fillna('', inplace=True)

# Assuming you have already preprocessed the text data and stored it in 'text'
text = news_df['text'].values

# TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf_vectorizer.fit_transform(text).toarray()

# Concatenate the TF-IDF features with the numerical features (tweet_num)
X = np.hstack((X_tfidf, news_df['tweet_num'].values.reshape(-1, 1)))

# Get feature names (including 'tweet_num')
feature_names = tfidf_vectorizer.get_feature_names_out() + ['tweet_num']

# Correlation matrix
correlation_matrix = np.corrcoef(X, rowvar=False)

# Extract correlation between 'real' (dependent variable) and other features
correlation_with_real = correlation_matrix[:-1, -1]

# Display correlation results
correlation_results = pd.DataFrame({'Feature': feature_names, 'Correlation with Real': correlation_with_real})
print(correlation_results)
```

# Feature Engineering
- Convert categorical or non-numeric features into a numerical representation
- Transform neccessary features using feature transformation techniques of your choice.

```python
# a function to convert 0 to fake and 1 to real in the real column
def convert_real(real_lst):
    real = []
    for val in real_lst:
        if val == 0:
            real.append('fake')
        elif val == 1:
            real.append('real')
    return real

# Apply the function to create a 'real' column
real = news_df['real'].tolist()
news_df['real'] = convert_real(real)
news_df
```

```python
news_df.iloc[23]['title']
```

```python
# Preprocesseing
import string
from nltk.tokenize import word_tokenize

# Preprocess text data
def preprocess_text(text):
     # Convert text to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove extra white spaces
    text = ' '.join(text.split())
    return text

# Apply text preprocessing to 'title' column
news_df['title'] = news_df['title'].apply(preprocess_text)

# Tokenize the text data
news_df['tags'] = news_df[['title', 'news_url', 'source_domain']].apply(lambda row: [word_tokenize(row['title']), word_tokenize(row['news_url']), word_tokenize(row['source_domain'])], axis=1)

# Combine tokens into a string (needed for feature extraction)
news_df['tags'] = news_df['tags'].apply(lambda tag: ' '.join(sum(tag, [])))

processed_news_df = news_df[['real', 'tags']]
processed_news_df
```

```python
processed_news_df.iloc[0]['tags']
```

# Machine Learning
- Use 2 different ML algorithms to build a model using your preprocessed data.
- Compare the 2 models based on their accuracy.

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier

# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(processed_news_df['tags'], 
                                                        processed_news_df['real'], test_size=0.2, random_state=42)

# Applied CountVectorizer
vectorizer = CountVectorizer(max_features=22800, stop_words='english')  # Limit to the top 22800 words
X_train_bow = vectorizer.fit_transform(X_train)
X_test_bow = vectorizer.transform(X_test)

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train_bow, y_train)
news_review_model = MultinomialNB()
news_review_model.fit(X_train_bow, y_train)
```

# Model Evaluation
- Evaluate the 2 models using a minimum of 4 evaluation metrics

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# analyse reviews using the model
prediction = news_review_model.predict(X_test_bow)

# Evaluate the model using different classification metrics
accuracy = accuracy_score(y_test, prediction)
classification_rep = classification_report(y_test, prediction)
conf_matrix = confusion_matrix(y_test, prediction)


# Print the results
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_rep)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train_bow, y_train, cv=10, scoring="accuracy")
```

```python
#Evaluate SGDClassifier
predict = sgd_clf.predict(X_test_bow)
accuracy = accuracy_score(y_test, predict)
classification_rep = classification_report(y_test, predict)
conf_matrix = confusion_matrix(y_test, predict)
# Print the results
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_rep)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier(random_state=42)
cross_val_score(forest_clf, X_train_bow, y_train, cv=3, scoring="accuracy")
```

```python
#Evaluate forest_clf
forest_clf.fit(X_train_bow, y_train)
predict = forest_clf.predict(X_test_bow)
accuracy = accuracy_score(y_test, predict)
classification_rep = classification_report(y_test, predict)
conf_matrix = confusion_matrix(y_test, predict)
# Print the results
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_rep)
print("Confusion Matrix:")
print(conf_matrix)
```
